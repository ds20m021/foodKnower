# Food identification Model

following model has been used:
https://huggingface.co/nateraw/food

This model tries to identify which food is visible on an image.
Base applikation runs in flask and a file has to be posted to the /predict endpoint. Returns the top 3 classes.

## Get it running
### Building the Docker file
* docker build -t food-knower .
* docker run -d -p 5000:5000 food-knower


### Pulling it from DockerHub
* docker pull ttasser/food-knower:v0.7.0
* docker run -d -p 5000:5000 ttasser/food-knower:v0.7.0

## Testing the service
The following code allows you to test the service, either use the locally running version by using the 127.0.0.1 url or the version running on azure.

```python
import requests
from IPython.display import Image
import pandas as pd
import json
import os
```

```python
service_url = "https://food-knower.azurewebsites.net/predict"
#service_url = "http://127.0.0.1:5000/predict"
```

```python
def test(filename):
    print(filename)
    display(Image(filename=filename, width=512,height=512))
    with open(filename, 'rb') as f:
        r = requests.post(service_url, files={'file': f})
        print(pd.DataFrame.from_dict(json.loads(r.text), orient='columns'))
```

```python
directory_list = os.listdir('testimages/')
for i in directory_list:
    test('testimages/'+i)
    print("--------------------------------------\n\n--------------------------------------")
```
